# Generate Synthetic Data

```python
import numpy as np

import os

import pandas as pd

import matplotlib.pyplot as plt
```

```python
save_path = 'data/'
N_sim = int(5e4)
sigma = 0.1
```

## $\mathbb{R}^{2}$ in $\mathbb{R}^{3}$

```python
path = os.path.join(save_path, 'R3/')
```

```python
if not os.path.exists(path):
    os.mkdir(path)
```

```python
path = os.path.join(path, 'data.csv')
```

```python
x1 = np.random.normal(0,1,N_sim)
x2 = np.random.normal(0,1,N_sim)
x3 = np.random.normal(0,sigma**2,N_sim)
```

```python
df = pd.DataFrame((x1,x2,x3))
```

```python
df.to_csv(path, index=False)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1,x2,x3)
ax.set_zlim(-2,2)
```

## $\mathbb{S}^{1}$ in $\mathbb{R}^{2}$

```python
path = os.path.join(save_path, 'CircleR2/')
```

```python
if not os.path.exists(path):
    os.mkdir(path)
```

```python
path = os.path.join(path, 'data.csv')
```

```python
theta = np.linspace(0,2*np.pi,N_sim)
x1 = np.cos(theta)+np.random.normal(0,sigma**2,N_sim)
x2 = np.sin(theta)+np.random.normal(0,sigma**2,N_sim)
```

```python
df = pd.DataFrame((x1,x2))
```

```python
df.to_csv(path, index=False)
```

```python
plt.plot(x1,x2)
```

## Hyperbolic Paraboloid in $\mathbb{R}^{3}$

```python
path = os.path.join(save_path, 'HypPar/')
```

```python
if not os.path.exists(path):
    os.mkdir(path)
```

```python
path = os.path.join(path, 'data.csv')
```

```python
x1 = np.random.normal(0,1,N_sim)
x2 = np.random.normal(0,1,N_sim)
x3 = x1**2-x2**2+np.random.normal(0,sigma**2,N_sim)
```

```python
df = pd.DataFrame((x1,x2,x3))
```

```python
df.to_csv(path, index=False)
```
